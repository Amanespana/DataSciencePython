```python
def get_relevant_skills(row_list):
    filtered_skills = []
    for skill in row_list:
        if skill in relevant_skills:
            filtered_skills.append(skill)
        else:
            continue
    return filtered_skills

def text_to_list(x):
    into_list = []
    x = x.lower()
    if '_x000d_' in x:
        into_list = x.split('_x000d_\n')
    else:
        into_list = x.split('\n')
        
    return into_list
```

```python
def label_for_syn(row_list):
    
    new_list = []
    for skill in row_list:
        #skill = skill.lstrip()
        for label, syn in syn_dict.items():
            if skill in syn:
                l=label
                break
            else:
                l = skill
        new_list.append(l)
    return new_list

syn_dict = {
    'ms sql'          : ['ms sql server', 'mssql', 'microsoft sql server', 'ms server'],
    'mysql'           : ['mysql server'],
    'azure'           : ['azure data explorer'],
    'oracle'          : ['oracle database'],
    'sap'             : ['sap hana'],
    'amazon dynamodb' : ['dynamodb'],
    'azure sql server': ['microsoft azure sql server'],
    'aws'             : ['qucksight', 'amazon quicksight', 'quick sight'],
    'sas'             : ['sas visual analytics'],
    'r'               : ['r programming', 'r language'],
    'spark'           : ['pyspark'],
    'sql'             : ['sql queries', 'sas sql', 'plsql', 'oracle plsql'],
    'excel'           : ['advanced excel', 'microsoft excel', 'ms excel', 'vba', 'macros', 'vba macros', 'microsoft office excel', 'pivot table', 'spreadsheet']

}
```

```python
def get_unique(row_list):
    unique_list = []
    for skill in row_list:
        if skill not in unique_list:
            unique_list.append(skill)
        else:
            continue

    return unique_list
```

```python
def get_stats(df, col_name):
    # getting stats df
    stats = pd.DataFrame(data = df.columns, columns = [col_name])
    
    #getting a pandas series
    ser = stats[col_name]
    
    #turning that series into a df showing value counts
    val_counts = ser.apply(lambda x: df[x].value_counts())
    
    #adding count col in final dfs
    stats['Count'] = val_counts[1]
    
    #adding percentage col to stats dfs
    stats['Percentage'] =  stats['Count'].apply(lambda x: 100*(x/len(df)))
    
    return stats
```

```python
#creating new df that will help to plot graphs
bi_stats = get_stats(bi_df, 'BI_Tool')
bi_stats
```
