```python
def get_relevant_skills(row_list):
    filtered_skills = []
    for skill in row_list:
        if skill in relevant_skills:
            filtered_skills.append(skill)
        else:
            continue
    return filtered_skills

def text_to_list(x):
    into_list = []
    x = x.lower()
    if '_x000d_' in x:
        into_list = x.split('_x000d_\n')
    else:
        into_list = x.split('\n')
        
    return into_list
```

```python
def get_stats(df, col_name):
    # getting stats df
    stats = pd.DataFrame(data = df.columns, columns = [col_name])
    
    #getting a pandas series
    ser = stats[col_name]
    
    #turning that series into a df showing value counts
    val_counts = ser.apply(lambda x: df[x].value_counts())
    
    #adding count col in final dfs
    stats['Count'] = val_counts[1]
    
    #adding percentage col to stats dfs
    stats['Percentage'] =  stats['Count'].apply(lambda x: 100*(x/len(df)))
    
    return stats
```
