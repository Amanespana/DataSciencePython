```python
relevant_skills= [ 'power bi', 'bi', 'sql', 'tableau', 'r', 'r programming', 'python', 'statistics', 'statistical analysis',
                 'oracle', 'sas', 'alteryx', 'sap', 'vba', 'vba macros', 'macros', 'hive', 'hadoop', 'spss', 'scala', 
                 'sql queries', 'machine learning', 'nosql', 'crm', 'data mining', 'excel', 'reports', 'dashboards', 
                 'data collection', 'business intelligence', 'sql server', 'database', 'databases', 'database design', 
                  'google analytics', 'google studio', 'google data studio', 'adobe analytics', 'advanced excel', 'mysql', 'mongodb', 
                 'snowflake', 'cloud', 'azure', 'aws', 'qlikview', 'pyspark', 'qliksense', 'qlik sense', 'ms sql', 
                 'big data', 'google sheets', 'r language', 'ms office', 'microsoft sql server', 'php', 'oracle database', 'plsql', 'oracle plsql',
                 'data extraction', 'microsoft office excel', 'sas sql']


def get_refined(s_text):
    refined_skills_scraped = []
    s_text = s_text.lower()
    skills_scraped_list = s_text.split(' ,')
    for skill in skills_scraped_list:
        skill = skill.lstrip()
        if skill in relevant_skills:
            refined_skills_scraped.append(skill)
        else:
            continue
    return refined_skills_scraped
```

```python
df['Skills'] = df['Skills'].apply(lambda x: get_refined(x))
```

```python
def get_text(row_list):
    filtered_skills = []
    for skill in row_list:
        if skill in relevant_skills:
            filtered_skills.append(skill)
        else:
            continue
    return filtered_skills
```

```python
def demo(x):
    into_list = []
    x = x.lower()
    if '_x000d_' in x:
        into_list = x.split('_x000d_\n')
    else:
        into_list = x.split('\n')
        
    return into_list
```
