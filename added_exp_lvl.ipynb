```python
def exp_lvl(x):
    if x <=3:
        lvl = 'Junior'
    elif x >3 and x<=5:
        lvl = 'Middle'
    elif x> 5 and x<=8:
        lvl = 'Senior'
    else:
        lvl = 'Executive'
    return lvl
```

```python
df_da['Experience Level'] = df_da['Experience Level'].apply(lambda x: exp_lvl(x))
```
